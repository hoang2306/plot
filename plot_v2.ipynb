```python
import matplotlib.pyplot as plt
import numpy as np

# dataset_name = 'Food'
# datasets = ['Baby', 'Sports', 'Clothing']
# datasets = ['Pog', 'Spotify']
# datasets = ['Pog']
# datasets = ['Spotify']
# datasets = ['Electronic'] 
# datasets = [dataset_name]
datasets = ['Pog', 'Spotify', 'Electronic', 'Food']

# categories = ['w/o text', 'w/o image', 'RaMen']
# categories = ['w/o CrSE', 'w/o CbSE', 'w/o ISL', 'w/o MAD', 'RaMen']
categories = ['w/o CrSE', 'w/o CbSE', 'w/o ISL', 'w/o MAD']

# values_recall = {
#     'Baby': [0.0417, 0.0591, 0.0771],
#     'Sports': [0.0544, 0.1041, 0.1152],
#     'Clothing': [0.0861, 0.1331, 0.1501]
# }
# values_ndcg = {
#     'Baby': [0.0195 , 0.0279, 0.0363],
#     'Clothing': [0.0394, 0.0661, 0.0729],
#     'Sports': [0.0266, 0.0513, 0.0574]
# }

# pog 
values_recall_pog = {
    # 'w/o CrSE', 'w/o CbSE', 'w/o ISL', 'w/o MAD'
    'Pog': [0.0218, 0.0301, 0.0335, 0.0348, 0.0375]
}
values_ndcg_pog = {
    'Pog': [0.0119, 0.0209, 0.0202, 0.0210, 0.0226]
}
values_recall_percent_pog = {
    'Pog': [41.87, 19.73, 10.67, 7.20]
}
values_ndcg_percent_pog = {
    'Pog': [47.35, 7.52, 10.62, 7.08]
}

# spotify 
values_recall_spotify = {
    # w/o text, w/o image, full model
    'Spotify': [0.1028, 0.0995, 0.1071, 0.1080, 0.1091]
}
values_ndcg_spotify = {
    'Spotify': [0.1772, 0.1662, 0.1863, 0.1838, 0.1882]
}
values_recall_percent_spotify = {
    'Spotify': [5.77, 8.8, 1.83, 1.01]
}
values_ndcg_percent_spotify = {
    'Spotify': [5.84, 11.69, 1.01, 2.34]
}

# electronic 
values_recall_electronic = {
    'Electronic': [0.7376, 0.3560, 0.8081, 0.7836, 0.8371]
}
values_ndcg_electronic = {
    'Electronic': [0.4780, 0.2595, 0.5253, 0.5035, 0.5373]
}
values_recall_percent_electronic = {
    'Electronic': [11.89, 57.47, 3.46, 6.39]
}
values_ndcg_percent_electronic = {
    'Electronic': [11.04, 51.7, 2.23, 6.29]
}

# food 
values_recall_food = {
    'Food': [0.7871, 0.3732, 0.8116, 0.8184, 0.8459]
}
values_ndcg_food = {
    'Food': [0.4815, 0.2788, 0.5122, 0.5104, 0.5242]
}
values_recall_percent_food = {
    'Food': [6.95, 55.88, 4.05, 3.25]
}
values_ndcg_percent_food = {
    'Food': [8.15, 46.81, 2.29, 2.63]
}


all_recall = {
    'Pog': values_recall_pog,
    'Spotify': values_recall_spotify,
    'Electronic': values_recall_electronic, 
    'Food': values_recall_food
}
all_recall_percent = {
    'Pog': values_recall_percent_pog,
    'Spotify': values_recall_percent_spotify,
    'Electronic': values_recall_percent_electronic, 
    'Food': values_recall_percent_food
}

all_ndcg = {
    'Pog': values_ndcg_pog,
    'Spotify': values_ndcg_spotify,
    'Electronic': values_ndcg_electronic,
    'Food': values_ndcg_food
}
all_ndcg_percent = {
    'Pog': values_ndcg_percent_pog,
    'Spotify': values_ndcg_percent_spotify,
    'Electronic': values_ndcg_percent_electronic,
    'Food': values_ndcg_percent_food
}

# data 
values_recall = all_recall_percent
values_ndcg = all_ndcg_percent

# Thit lp v tr ct
x = np.arange(len(datasets))
width = 0.15  # width of cols
# spacing = 0.06 # 
spacing = 0.03 # space between cols

# V biu  Recall
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(22, 15))
offset = -((len(categories) - 1) * (width + spacing) / 2)
rects_recall = []
rects_ndcg = []
# colors = ['#A6CEE3', '#B2DF8A', '#FDBF6F']  # 
colors = ['#A6CEE3', '#B2DF8A', '#9299d1', '#FDBF6F', '#FDBF6F']  
hatch_patterns = ['/', 'x', '\\', 'xx', '']  # 

for i, category in enumerate(categories):
    values_list_recall = [values_recall[dataset][dataset][i] for dataset in datasets]
    rects_recall.append(ax1.bar(x + offset + i * (width + spacing), values_list_recall, width, label=category, color=colors[i], hatch=hatch_patterns[i], edgecolor='black'))
    
    values_list_ndcg = [values_ndcg[dataset][dataset][i] for dataset in datasets]
    rects_ndcg.append(ax2.bar(x + offset + i * (width + spacing), values_list_ndcg, width, label=category, color=colors[i], hatch=hatch_patterns[i], edgecolor='black'))

# 
ax1.yaxis.grid(True, linestyle='--', linewidth=1, alpha=1)
ax1.set_axisbelow(True)
ax2.yaxis.grid(True, linestyle='--', linewidth=1, alpha=1)
ax2.set_axisbelow(True)

# Gn nhn v tiu 
# ax1.set_ylabel('R@20', fontsize=12)
ax1.set_title(r'% Reduction at Recall@20', fontsize=20, fontweight='bold')
ax1.set_xticks(x)

# if dataset_name == "Pog":
#     ax1.set_ylim(0, 0.045)
# if dataset_name == 'Spotify':
#     ax1.set_ylim(0, 0.13)
# if dataset_name == 'Electronic':
#     ax1.set_ylim(0, 1)
# if dataset_name == 'Food':
#     ax1.set_ylim(0, 1)
ax1.set_ylim(0, 68)

ax1.set_xticklabels(datasets, fontsize=20)
# ax1.legend(loc="center", title="", fontsize=13, bbox_to_anchor=(0.5, -0.2), ncol=3)
# ax1.set_xlabel('a)')

# ax2.set_ylabel('NDCG@20', fontsize=12)
# ax2.set_title('NDCG@20', fontsize=13)
ax2.set_title(r'% Reduction at NDCG@20', fontsize=20, fontweight='bold')
ax2.set_xticks(x)

# if dataset_name == "Pog":
#     ax2.set_ylim(0, 0.027)
# if dataset_name == "Spotify":
#     ax2.set_ylim(0, 0.22)
# if dataset_name == 'Electronic':
#     ax2.set_ylim(0, 0.6)
# if dataset_name == 'Food':
#     ax2.set_ylim(0, 0.6)

ax2.set_ylim(0, 60)
ax2.set_xticklabels(datasets, fontsize=20)
# ax2.legend(title="", fontsize=13)
# ax2.set_xlabel('b)')


# set legend for fig 
fig.legend(categories, loc="center", title="", fontsize=20, bbox_to_anchor=(0.5, -0.02), ncol=5)

# display value in cols
for rect_group in rects_recall:
    for rect in rect_group:
        height = rect.get_height()
        ax1.annotate(f'{height:.2f}%',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=15, fontweight='bold')

for rect_group in rects_ndcg:
    for rect in rect_group:
        height = rect.get_height()
        ax2.annotate(f'{height:.2f}%',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=15, fontweight='bold')

plt.tight_layout()
plt.savefig('./plot.pdf')
plt.show()
```
